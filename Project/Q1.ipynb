```python
import random
import math as m
from matplotlib import pyplot as plt
```

# functions


```python
def mean(list):
    sum = 0.0
    
    for i in list:
        sum = sum + i 
        
    return sum/len(list)

def rms(list):
    meanOfList = mean(list)
    Dsum = 0.0
    
    for i in list:
        Dsum = Dsum + (meanOfList - i)**2
        
    return m.sqrt(Dsum/len(list))
        
    
        
    
        


def walk(step_size):
    points = [(0,0)]
    for i in range(0,step_size):
        dx = m.sin(random.uniform(0,1) * m.pi * 2)
        dy = m.cos(random.uniform(0,1) * m.pi * 2)
        newPoint = (points[-1][0]+ dx,points[-1][1]+ dy)
        points.append(newPoint)
            
    return points

def walkClass(step_size):
    walks = []
    Dists = []
    XDists = []
    YDists = []
    
    for i in range(0,200):
        walks.append(walk(step_size))
    
    for i in range(0,200):
        #print(i)
        #print(walks[i][-1][0]**2 + walks[i][-1][1]**2)
        Dist = m.sqrt(walks[i][-1][0]**2 + walks[i][-1][1]**2) 
        XDist = walks[i][-1][0]
        YDist = walks[i][-1][1]
        #print(meanDist)
        Dists.append(Dist)
        XDists.append(XDist)
        YDists.append(YDist)
        
        
    
    return walks[0:5],mean(Dists),rms(Dists),mean(XDists),rms(XDists),mean(YDists),rms(YDists)


steps = [250,500,750,1000,1250]
#steps = [250]

dataRmsVsSqN =[] 
for step in steps:
    walks,meanDist,rmsDist,meanXDist,rmsXDist,meanYDist,rmsYDist = walkClass(step)
    dataRmsVsSqN.append((m.sqrt(step),rmsDist))
    print("Mean Distance for step size ",i," is ",meanDist,"with RMS ",rmsDist,".")
    print("Mean X directin Displacement is ", meanXDist," with RMS",rmsXDist,"." )
    print("Mean Y direction Displacement is ", meanYDist," with RMS",rmsYDist,".")
    for data in walks:
        plt.plot(*zip(*data))
    plt.grid(True)
    plt.show()

plt.plot(*zip(*dataRmsVsSqN))
plt.grid(True)
plt.show
```
